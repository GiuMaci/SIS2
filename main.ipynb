```python
sys.path = original_path

from src.evaluator import computator
from src.core import reader, onto_dict, split_by_caps,embedded_dict,myembed,matchrelated,topN
from src.core import remove_special_characters, matcher_of_indexes, RTXmatcher,SEMEMBmatcher
```

```python
import pandas as pd

datas=pd.DataFrame()
for al in names:
    print(al)
    name1=al.split('-')[0]
    name2=al.split('-')[1]
    cmt=reader(absolute_path+'/'+name1+'.txt')
    ekaw=reader(absolute_path+'/'+name2+'.txt')
    cmt_dict=onto_dict(cmt)
    ekaw_dict=onto_dict(ekaw)
    cmtRTX=[]
    ekawRTX=[]
    for k in range(0,len(cmt)):
        string=split_by_caps(remove_special_characters(cmt[k][0]+cmt[k][1]+cmt[k][2]))
        cmtRTX.append(string)
    for k in range(0,len(ekaw)):
        string=split_by_caps(remove_special_characters(ekaw[k][0]+ekaw[k][1]+ekaw[k][2]))
        ekawRTX.append(string)
    embedded_dict_cmt=embedded_dict(cmt_dict)
    embedded_dict_ekaw=embedded_dict(ekaw_dict)
    cmtvec,ekawvec=myembed(cmtRTX,ekawRTX,0)
    Ks=[1,2,3,5,8,10,12,15,18,25]
    for K in Ks:
        print(K)
        RTXmatcher(K,cmt,ekaw,cmtvec,ekawvec,os.getcwd()+str(al)+".csv")
    #MATCHING CON SOLO RELATED TXT
        result=computator(os.getcwd()+str(al)+".csv",sys.path[0]+'/src/alignment/'+al+"-referencealignment.csv")
        #datas[al+'RELTX']=None
        datas.loc[K,al+'RELTX']=result[0]
    
    # MATCHING PIU GENERALE
        SEMEMBmatcher(K,cmt,ekaw,embedded_dict_cmt,embedded_dict_ekaw,os.getcwd()+str(al)+"SEMEMB.csv")
        result=computator(os.getcwd()+str(al)+"SEMEMB.csv",sys.path[0]+'/src/alignment/'+al+"-referencealignment.csv")
        #datas[al+'SEMEMB']=None
        datas.loc[K,al+'SEMEMB']=result[0]
datas.round(2)
datas.to_csv('final_eval_output.csv', index=False)
```
